```python
tagger = StanfordNERTagger("/home/ikscare/Documents/Projects/Mousam/stanford-ner-2014-08-27/classifiers/english.all.3class.distsim.crf.ser.gz",
                       path_to_jar="/home/ikscare/Documents/Projects/Mousam/stanford-ner-2014-08-27/stanford-ner.jar",
                          encoding='utf-8')
```

```python
tokenized_paragraphs=[]
for sentence in sentences:
    tokenized_paragraphs.append(word_tokenize(sentence))

classified_paragraphs_list = tagger.tag_sents(tokenized_paragraphs)
```

```python
persons=[]
organistions=[]
locations=[]
temp_entity_name = ''
```

```python
for classified_paragraph in classified_paragraphs_list:
    for entry in classified_paragraph:
        entry_value = entry[0]
        entry_type = entry[1]
        if entry_type == 'PERSON':
            temp_entity_name = ' '.join([temp_entity_name, entry_value]).strip()
            persons.append(temp_entity_name)
        elif entry_type == 'ORGANIZATION':
            organizations.append(entry_value)
            temp_entity_name = ''
        elif entry_type == 'LOCATION':
            locations.append(entry_value)
        temp_entity_name = ''
```
