```python
#document="""Hello this is Mousam Singh
#How are you
#Is't it a good touch"""
sentences=document.split('\n')
```

```python
import spacy
import pandas as pd
#sentences=document.split('\n')
#nlp = spacy.load('en_core_web_sm')
nlp = spacy.load('en')
named_entities = []
for sentence in sentences:
    #print(sentence)
    temp_entity_name = ''
    temp_named_entity = None
    sentence = nlp(sentence)
    for word in sentence:
        #print(word)
        term = word.text 
        tag = word.ent_type_
        
        if tag:
            #print(tag)
            temp_entity_name = ' '.join([temp_entity_name, term]).strip()
            temp_named_entity = (temp_entity_name, tag)
            
        else:
            #print(temp_named_entity)
            if temp_named_entity:
                print('In inner if',temp_named_entity)
                named_entities.append(temp_named_entity)
                temp_entity_name = ''
                temp_named_entity = None

entity_frame = pd.DataFrame(set(named_entities), 
                            columns=['Entity Name', 'Entity Type'])
```
